```python
def getData(data_folder):
    """
    Walk through a folder, appending to a list only MID files
    :param data_folder: the base data folder used in the model, still nonrepresentative
    :return data_frame: a df with all data that can be extracted from class mido.MidiFile
    """
    data_frame = pd.DataFrame()
    
    # used to clean the folder after running for the first time
    ignored = 0
    i=0
    
    for dirpath, _, filenames in os.walk(data_folder):
        for File in filenames:
            path = os.path.join(dirpath, File)
            
            if not (path.endswith('.mid') or path.endswith('.midi')):
                print(f"{path} ignored")
                ignored +=1
                continue

            try:
                mid = mido.MidiFile(path)
                tpb = mid.ticks_per_beat
                duration = mid.length
                """type 0 (single track): all messages are saved in one track
                type 1 (synchronous): all tracks start at the same time
                type 2 (asynchronous): each track is independent of the others"""
                mid_type = mid.type        
                

            except:
                os.remove(path)
                re+=1
                continue
        
            data_frame = data_frame.append({"tick_beat":tpb, "duration":round(duration,2), "type": mid_type}, ignore_index = True)
            
            i+=1
            if i % 100 == 0:
                print(i)
    return data_frame
```
